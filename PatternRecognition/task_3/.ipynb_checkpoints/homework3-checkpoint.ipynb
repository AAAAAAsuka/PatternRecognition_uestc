```python
import math
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from scipy import io
from sklearn import datasets
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import PolynomialFeatures,StandardScaler,MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from sklearn.svm import SVC,LinearSVC
```

```python
# 数据读取
train_data = io.loadmat('train_data.mat')['train_data']
train_label = io.loadmat('train_label.mat')['train_label'].reshape(-1)
test_data = io.loadmat('test_data.mat')['test_data']
# pos 120 neg 300
```

```python
# 绘图函数
def plot_faces(faces):
    fig, axes = plt.subplots(6, 6, figsize=(10,10),
                            subplot_kw ={'xticks':[], 'yticks':[]},
    gridspec_kw = dict(hspace=0.1,wspace=0.1))
    for i, ax in enumerate(axes.flat):
        ax.imshow(faces[i].reshape(19,19,order = 'F'),cmap = 'bone')
        plt.show
        
# 查看图像       
plot_faces(train_data[:36])
plot_faces(train_data[-36:])
plot_faces(test_data[:36])
plot_faces(test_data[-36:])
```

```python
len(train_data),train_data.shape
```

```python
# 图像增强

def calcGrayHist(I):
    # 计算灰度直方图
    h, w = I.shape[:2]
    grayHist = np.zeros([256], np.uint64)
    for i in range(h):
        for j in range(w):
#             print(I)
            grayHist[int(I[i][j])] += 1
    return grayHist

def equalHist(img):
    # 灰度图像矩阵的高、宽
    h, w = img.shape
    # 第一步：计算灰度直方图
    grayHist = calcGrayHist(img)
    # 第二步：计算累加灰度直方图
    zeroCumuMoment = np.zeros([256], np.uint32)
    for p in range(256):
        if p == 0:
            zeroCumuMoment[p] = grayHist[0]
        else:
            zeroCumuMoment[p] = zeroCumuMoment[p - 1] + grayHist[p]
    # 第三步：根据累加灰度直方图得到输入灰度级和输出灰度级之间的映射关系
    outPut_q = np.zeros([256], np.uint8)
    cofficient = 256.0 / (h * w)
    for p in range(256):
        q = cofficient * float(zeroCumuMoment[p]) - 1
        if q >= 0:
            outPut_q[p] = math.floor(q)
        else:
            outPut_q[p] = 0
    # 第四步：得到直方图均衡化后的图像
    equalHistImage = np.zeros(img.shape, np.uint8)
    for i in range(h):
        for j in range(w):
            equalHistImage[i][j] = outPut_q[int(img[i][j])]
    return equalHistImage

def hist_norm(data):
    output_data = []
    img = data.reshape((len(data),19, 19))*255
    for k in range(len(data)): 
        out = equalHist(img[k])
        output_data.append(out)
    return np.array(output_data).reshape((len(data),-1))/255

# 直方图均衡化
train_data_norm = hist_norm(train_data)
test_data_norm = hist_norm(test_data)

# 均衡化后图像
plot_faces(train_data_norm[:36])
plot_faces(train_data_norm[-36:])
plot_faces(test_data_norm[:36])
plot_faces(test_data_norm[-36:])
```

```python
# 图像旋转
def pic_flip(data):
    output_data0 = []
    output_data1 = []
    img = data.reshape((len(data),19, 19))
    for k in range(len(data)): 
        out0 = np.flip(img[k], axis = 0)
        output_data0.append(out0)
        out1 = np.flip(img[k], axis = 1)
        output_data1.append(out1)
    return np.array(output_data0).reshape((len(data),-1)), np.array(output_data1).reshape((len(data),-1))

train_data_norm_pos, train_data_norm_neg = train_data_norm[:400], train_data_norm[400:]

neg0, neg1 = pic_flip(train_data_norm_neg)
train_data_norm_fliped_neg = np.concatenate((train_data_norm, neg0, neg1), axis = 0)
train_label_fliped_neg = np.concatenate((train_label, np.ones([800])* -1), axis = 0)
print(train_data_norm_fliped_neg.shape)

pos0, _ = pic_flip(train_data_norm_pos)
train_data_norm_fliped_pos = np.concatenate((pos0, train_data_norm), axis = 0)
train_label_fliped_pos = np.concatenate((np.ones([400]), train_label), axis = 0)
print(train_data_norm_fliped_pos.shape)

train_data_norm_fliped = np.concatenate((pos0, train_data_norm, neg0, neg1), axis = 0)
train_label_fliped = np.concatenate((np.ones([400]), train_label, np.ones([800])* -1), axis = 0)
print(train_data_norm_fliped.shape)
```

```python
plot_faces(train_data_norm_fliped[:36])
plot_faces(train_data_norm_fliped[-36:])
plot_faces(train_data_norm_fliped[-436:])
plot_faces(train_data_norm_fliped[-836:])
```

```python
# 归一化
from sklearn.preprocessing import StandardScaler

standardScaler = StandardScaler()
standardScaler.fit(X)
X_standard = standardScaler.transform(X)
```

```python
def plot_decision_boundary(model, axis):
    
    x0, x1 = np.meshgrid(
        np.linspace(axis[0], axis[1], int((axis[1]-axis[0])*100)).reshape(-1, 1),
        np.linspace(axis[2], axis[3], int((axis[3]-axis[2])*100)).reshape(-1, 1),
    )
    X_new = np.c_[x0.ravel(), x1.ravel()]

    y_predict = model.predict(X_new)
    zz = y_predict.reshape(x0.shape)

    from matplotlib.colors import ListedColormap
    custom_cmap = ListedColormap(['#EF9A9A','#FFF59D','#90CAF9'])
    
    plt.contourf(x0, x1, zz, linewidth=5, cmap=custom_cmap)
    
def plot_svc_decision_boundary(model, axis):
    
    x0, x1 = np.meshgrid(
        np.linspace(axis[0], axis[1], int((axis[1]-axis[0])*100)).reshape(-1, 1),
        np.linspace(axis[2], axis[3], int((axis[3]-axis[2])*100)).reshape(-1, 1),
    )
    X_new = np.c_[x0.ravel(), x1.ravel()]

    y_predict = model.predict(X_new)
    zz = y_predict.reshape(x0.shape)

    from matplotlib.colors import ListedColormap
    custom_cmap = ListedColormap(['#EF9A9A','#FFF59D','#90CAF9'])
    
    plt.contourf(x0, x1, zz, linewidth=5, cmap=custom_cmap)
    
    w = model.coef_[0]
    b = model.intercept_[0]
    
    # w0*x0 + w1*x1 + b = 0
    # => x1 = -w0/w1 * x0 - b/w1
    plot_x = np.linspace(axis[0], axis[1], 200)
    up_y = -w[0]/w[1] * plot_x - b/w[1] + 1/w[1]
    down_y = -w[0]/w[1] * plot_x - b/w[1] - 1/w[1]
    
    up_index = (up_y >= axis[2]) & (up_y <= axis[3])
    down_index = (down_y >= axis[2]) & (down_y <= axis[3])
    plt.plot(plot_x[up_index], up_y[up_index], color='black')
    plt.plot(plot_x[down_index], down_y[down_index], color='black')
```

```python
svc9 = LinearSVC(C=1e9)
svc9.fit(X_standard, y)

plot_svc_decision_boundary(svc9, axis=[-3, 3, -3, 3])
plt.scatter(X_standard[y==0,0], X_standard[y==0,1])
plt.scatter(X_standard[y==1,0], X_standard[y==1,1])
plt.show()
```

```python
X, y = datasets.make_moons(noise=0.15, random_state=666)

plt.scatter(X[y==0,0], X[y==0,1])
plt.scatter(X[y==1,0], X[y==1,1])
plt.show()
```

```python
# 多项式特征
def PolynomialSVC(degree, C=1.0):
    return Pipeline([
        ("poly", PolynomialFeatures(degree=degree)),
        ("std_scaler", StandardScaler()),
        ("linearSVC", LinearSVC(C=C))
    ])
```

```python
poly_svc = PolynomialSVC(degree=7, C = 1e7)
poly_svc.fit(X, y)

plot_decision_boundary(poly_svc, axis=[-1.5, 2.5, -1.0, 1.5])
plt.scatter(X[y==0,0], X[y==0,1])
plt.scatter(X[y==1,0], X[y==1,1])
plt.show()
```

```python
# 多项式核
def PolynomialKernelSVC(degree, C=1.0):
    return Pipeline([
        ("std_scaler", StandardScaler()),
        ("kernelSVC", SVC(kernel="poly", degree=degree, C=C))
    ])

poly_kernel_svc = PolynomialKernelSVC(degree=1,C=0.001)
poly_kernel_svc.fit(X, y)


plot_decision_boundary(poly_kernel_svc, axis=[-1.5, 2.5, -1.0, 1.5])
plt.scatter(X[y==0,0], X[y==0,1])
plt.scatter(X[y==1,0], X[y==1,1])
plt.show()
```

```python
# rbf核
def RBFKernelSVC(gamma):
    return Pipeline([
        ("std_scaler", StandardScaler()),
        ("svc", SVC(kernel="rbf", gamma=gamma))
    ])

svc_gamma1 = RBFKernelSVC(gamma=1)
svc_gamma1.fit(X, y)
```

```python
plot_decision_boundary(svc_gamma1, axis=[-1.5, 2.5, -1.0, 1.5])
plt.scatter(X[y==0,0], X[y==0,1])
plt.scatter(X[y==1,0], X[y==1,1])
plt.show()
```
